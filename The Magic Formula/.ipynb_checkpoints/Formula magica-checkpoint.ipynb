```python
# Carrega a base de dados de ROIC consolidado por ativo
df_ROIC = pd.read_excel("dados_consolidados_ROIC.xlsx", index_col="Ativo")
df_ROIC.head()
```

```python
# Carrega a base de dados de EV/EBIT consolidado por ativo
df_EV_EBIT = pd.read_excel("dados_consolidados_EV_EBIT.xlsx", index_col="Ativo")
df_EV_EBIT.head()
```

# Tratamento dos dados

#### Return on Invested Capital (ROIC)

```python
# =============================
# Pré-processamento do DataFrame ROIC
# =============================

# Inverte a ordem das colunas para que os anos fiquem do mais antigo (esquerda) para o mais recente (direita)
df_ROIC = df_ROIC[df_ROIC.columns[::-1]]

# Padroniza os nomes dos ativos para letras maiúsculas
df_ROIC.index = df_ROIC.index.str.upper()

# Move o índice ("Ativo") para uma coluna normal
df_ROIC = df_ROIC.reset_index()

# Transforma o DataFrame de formato "largo" para "longo":
# De: colunas com anos (2018, 2019, 2020...)
# Para: colunas "Ano" e "Indicador", onde cada linha representa o ROIC de um ativo em um ano
df_ROIC = df_ROIC.melt(id_vars="Ativo",
                       var_name="Ano",
                       value_name="Indicador")

# Remove manualmente a empresa PORT3 dos anos em que ela não possui dados confiáveis
df_ROIC = df_ROIC[~((df_ROIC["Ativo"] == "PORT3") & (df_ROIC["Ano"].isin([2018, 2019, 2020])))]
```

```python
# =============================
# Limpeza dos dados e cálculo do ranking (ROIC)
# =============================

# Converte os valores da coluna "Indicador" para string, remove espaços e substitui vírgulas por ponto decimal
df_ROIC["Indicador"] = df_ROIC["Indicador"].astype(str).str.strip().str.replace(",", ".")

# Remove qualquer caractere que não seja número, ponto ou sinal de negativo
# Isso elimina %, textos ou símbolos que possam ter vindo na planilha
df_ROIC["Indicador"] = df_ROIC["Indicador"].str.replace(r"[^\d\.\-]", "", regex=True)

# Converte os valores da coluna "Indicador" para tipo numérico (float), usando NaN em caso de erro
df_ROIC["Indicador"] = pd.to_numeric(df_ROIC["Indicador"], errors="coerce")

# Remove linhas com valores ausentes ou inválidos na coluna "Indicador"
df_ROIC = df_ROIC.dropna(subset=["Indicador"])

# Calcula o ranking do ROIC para cada ano (quanto maior o ROIC, melhor o ranking)
df_ROIC["Ranking"] = df_ROIC.groupby("Ano")["Indicador"].rank(ascending=False, method="min")

# Converte o ranking para tipo inteiro (permite valores nulos com Int64)
df_ROIC["Ranking"] = df_ROIC["Ranking"].astype("Int64")

# Ordena o DataFrame por Ano e Ranking (do melhor para o pior por ano)
df_ROIC = df_ROIC.sort_values(by=["Ano", "Ranking"]).reset_index(drop=True)

# Visualiza as primeiras linhas do DataFrame df_ROIC
df_ROIC.head()
```

#### Earning Yield (EBIT/EV)

```python
# =============================
# Pré-processamento do DataFrame EV/EBIT
# =============================

# Inverte a ordem das colunas para que os anos fiquem do mais antigo (esquerda) para o mais recente (direita)
df_EV_EBIT = df_EV_EBIT[df_EV_EBIT.columns[::-1]]

# Padroniza os nomes dos ativos para caixa alta (letras maiúsculas)
df_EV_EBIT.index = df_EV_EBIT.index.str.upper()

# Move o índice ("Ativo") para uma coluna normal
df_EV_EBIT = df_EV_EBIT.reset_index()

# Transforma o DataFrame de formato "largo" para "longo":
# De: colunas com anos (2018, 2019, 2020...)
# Para: colunas "Ano" e "Indicador", onde cada linha representa o EV/EBIT de um ativo em um ano
df_EV_EBIT = df_EV_EBIT.melt(id_vars="Ativo", 
                             var_name="Ano", 
                             value_name="Indicador")
```

```python
# =============================
# Limpeza, transformação e cálculo do ranking (EY)
# =============================

# Converte os valores da coluna "Indicador" para string, remove espaços e substitui vírgula por ponto decimal
df_EV_EBIT["Indicador"] = df_EV_EBIT["Indicador"].astype(str).str.strip().str.replace(",", ".")

# Remove qualquer caractere que não seja número, ponto ou sinal de negativo
df_EV_EBIT["Indicador"] = df_EV_EBIT["Indicador"].str.replace(r"[^\d\.\-]", "", regex=True)

# Converte os valores da coluna "Indicador" para tipo numérico (float), tratando erros como NaN
df_EV_EBIT["Indicador"] = pd.to_numeric(df_EV_EBIT["Indicador"], errors="coerce")

# Cria uma cópia para calcular EY (Earnings Yield = 1 / EV/EBIT)
df_EY = df_EV_EBIT.copy()

# Aplica a inversão: EY = 1 / (EV/EBIT), tratando divisões por zero e valores nulos
df_EY["Indicador"] = df_EV_EBIT["Indicador"].apply(lambda x: 1 / x if pd.notnull(x) and x != 0 else None)

# Remove linhas com valores inválidos ou não calculáveis
df_EY = df_EY.dropna(subset=["Indicador"])

# Calcula o ranking do EY para cada ano (quanto maior o EY, melhor o ranking)
df_EY["Ranking"] = df_EY.groupby("Ano")["Indicador"].rank(ascending=False, method="min")

# Converte o ranking para tipo inteiro, permitindo valores nulos com Int64
df_EY["Ranking"] = df_EY["Ranking"].astype("Int64")

# Ordena o DataFrame para facilitar análise posterior
df_EY = df_EY.sort_values(by=["Ano", "Ranking"]).reset_index(drop=True)

# Visualiza as primeiras linhas do DataFrame df_EY
df_EY.head()
```

# Backtesting estratégia

#### Criação de funções necessárias à avaliação

```python
def const_port_FM(df_ROIC, df_EY, n_ativos):
    """
    Constroi e simula a carteira da Fórmula Mágica para o mercado brasileiro.

    Parâmetros:
        df_ROIC (DataFrame): DataFrame com ranking de ROIC por ativo e ano
        df_EY (DataFrame): DataFrame com ranking de EY por ativo e ano
        n_ativos (int): Número de ativos a serem incluídos em cada carteira anual

    Retorna:
        DataFrame com retornos diários e acumulados da carteira e do IBOV
    """

    # Junta os dois rankings com base em "Ativo" e "Ano"
    df = pd.merge(
        df_ROIC[["Ativo", "Ano", "Ranking"]],
        df_EY[["Ativo", "Ano", "Ranking"]],
        on=["Ativo", "Ano"],
        suffixes=("_ROIC", "_EY")
    )

    # Calcula o ranking total como a soma dos rankings individuais (quanto menor, melhor)
    df["Ranking_Total"] = df["Ranking_ROIC"] + df["Ranking_EY"]

    resultados = []  # Lista para armazenar os resultados de cada ano
    capital_acumulado = 1  # Começa com 1 para calcular o retorno acumulado da carteira
    capital_acumulado_ibov = 1  # Idem para o IBOV
    anos = sorted(df["Ano"].unique())  # Lista de anos disponíveis

    # Loop por cada ano para montar a carteira
    for ano in anos:

        # Seleciona os top N ativos com melhor colocação no ranking total para aquele ano
        df_ano = df[df["Ano"] == ano].sort_values("Ranking_Total").head(n_ativos)
        ativos = df_ano["Ativo"].tolist()

        # Formata os tickers para o padrão do Yahoo Finance (.SA para ações brasileiras)
        tickers = [ativo if "." in ativo else ativo + ".SA" for ativo in ativos]

        # Define o período de backtest: ano seguinte à seleção
        start = f"{ano + 1}-01-01"
        end = f"{ano + 1}-12-31"

        try:
            # Baixa preços de fechamento dos ativos e do IBOV
            df_precos = yf.download(tickers, start=start, end=end, progress=False)["Close"]
            time.sleep(1)  # Pausa para evitar limite de requisições
            ibov = yf.download("^BVSP", start=start, end=end,progress=False)["Close"]
            time.sleep(2)
        except Exception as e:
            print(f"⚠️  Erro ao baixar dados para o ano {ano + 1}: {e}")
            continue

        # Verifica se dados foram baixados com sucesso
        if df_precos.empty or ibov.empty:
            print(f"⚠️  Dados não disponíveis para o ano {ano + 1}. Pulando...")
            continue

        # Calcula o retorno diário e acumulado da carteira (igual ponderação entre ativos)
        df_retorno = df_precos.pct_change().dropna()
        df_retorno["Retorno Diário"] = df_retorno.mean(axis=1)
        df_retorno["Retorno Acumulado"] = (1 + df_retorno["Retorno Diário"]).cumprod() * capital_acumulado
        capital_acumulado = df_retorno["Retorno Acumulado"].iloc[-1]  # Atualiza capital para o próximo ano

        # Processa o retorno do IBOV no mesmo período
        ibov.index = pd.to_datetime(ibov.index)
        ibov = ibov.reindex(df_retorno.index)
        ibov_ret_diario = ibov.pct_change().fillna(0)
        ibov_ret_acumulado = (1 + ibov_ret_diario).cumprod() * capital_acumulado_ibov
        capital_acumulado_ibov = ibov_ret_acumulado.iloc[-1]

        # Organiza os resultados do ano em um DataFrame
        df_resultado = df_retorno[["Retorno Diário", "Retorno Acumulado"]].copy()
        df_resultado["Data"] = df_resultado.index
        df_resultado["Ano Base"] = ano
        df_resultado["Ativos"] = ", ".join(ativos)

        # Adiciona os dados do IBOV alinhados às datas da carteira
        df_resultado["Pontos IBOV"] = ibov.reindex(df_resultado["Data"]).values
        df_resultado["Retorno Diário IBOV"] = ibov_ret_diario.reindex(df_resultado["Data"]).fillna(0).values
        df_resultado["Retorno Acumulado IBOV"] = ibov_ret_acumulado.reindex(df_resultado["Data"]).fillna(method="ffill").values

        # Armazena os resultados deste ano
        resultados.append(df_resultado)

    # Garante que pelo menos um ano foi processado com sucesso
    if not resultados:
        raise ValueError("Nenhum dado de retorno foi gerado. Verifique se os tickers e datas estão corretos.")

    # Concatena todos os anos em um único DataFrame final
    df_final = pd.concat(resultados).reset_index(drop=True)
    
    return df_final[[ 
        "Data", "Ano Base", "Ativos",
        "Retorno Diário", "Retorno Acumulado",
        "Pontos IBOV", "Retorno Diário IBOV", "Retorno Acumulado IBOV"
    ]]
```

#### Resultados

```python
def backtesting_FM(df_ROIC, df_EY, lista_n_ativos):
    """
    Realiza backtest da Fórmula Mágica para diferentes tamanhos de carteira.
    Exibe um gráfico estilizado com destaque para a crise de 2020 e retornos finais anotados.

    Parâmetros:
        df_ROIC (DataFrame): Ranking de ROIC por ativo e ano
        df_EY (DataFrame): Ranking de Earnings Yield por ativo e ano
        lista_n_ativos (list of int): Lista com diferentes quantidades de ativos a testar (ex: [5, 10, 20])

    Retorna:
        Nenhum. Apenas exibe o gráfico comparativo.
    """

    # Função interna para formatar os trimestres no eixo X
    def formatar_trimestre(x, pos):
        data = mdates.num2date(x)
        ano = str(data.year)[-2:]  # Pega apenas os dois últimos dígitos do ano
        mes = data.month
        trimestre = (mes - 1) // 3 + 1
        return f"{trimestre}T{ano}"

    fig, ax = plt.subplots(figsize=(14, 7))
    resultados = []

    for n in lista_n_ativos:
        try:
            df_resultado = const_port_FM(df_ROIC, df_EY, n_ativos=n)
            resultados.append((n, df_resultado))

            # Plot da carteira
            ax.plot(
                df_resultado["Data"],
                df_resultado["Retorno Acumulado"],
                label=f"Carteira FM - {n} ativos"
            )

            # Anotação do retorno final da carteira
            valor_final = df_resultado["Retorno Acumulado"].iloc[-1]
            data_final = df_resultado["Data"].iloc[-1]
            ax.annotate(f"{valor_final:.2f}x",
                        xy=(data_final, valor_final),
                        xytext=(5, 0),
                        textcoords="offset points",
                        fontweight="bold",
                        fontsize=10,
                        va="center",
                        color="black")

        except Exception as e:
            print(f"Erro ao processar {n} ativos: {e}")

    # Plot do IBOV (usando último resultado válido)
    if resultados:
        _, df_resultado = resultados[-1]

        ax.plot(
            df_resultado["Data"],
            df_resultado["Retorno Acumulado IBOV"],
            label="IBOV",
            linestyle="--",
            color="black",
            linewidth=1.5,
            dashes=(6, 4)
        )

        # Anotação final do IBOV
        valor_ibov_final = df_resultado["Retorno Acumulado IBOV"].iloc[-1]
        data_ibov_final = df_resultado["Data"].iloc[-1]
        ax.annotate(f"{valor_ibov_final:.2f}x",
                    xy=(data_ibov_final, valor_ibov_final),
                    xytext=(5, 0),
                    textcoords="offset points",
                    fontweight="bold",
                    fontsize=10,
                    va="center",
                    color="black")

    # Destaque da crise COVID-19
    pandemia_inicio = datetime(2020, 2, 20)
    pandemia_fim = datetime(2020, 4, 10)
    pandemia_meio = pandemia_inicio + (pandemia_fim - pandemia_inicio) / 2

    ax.axvspan(pandemia_inicio, pandemia_fim, color='lightgray', alpha=0.5)
    ax.text(pandemia_meio, ax.get_ylim()[1] * 0.90,
            "COVID-19", color="red", fontsize=8, fontweight="bold",
            ha="center")

    # Eixo X: aplicar formatador de trimestre
    ax.xaxis.set_major_locator(mdates.MonthLocator(interval=3))
    ax.xaxis.set_major_formatter(plt.FuncFormatter(formatar_trimestre))
    plt.setp(ax.get_xticklabels(), rotation=0, ha="center", fontsize=9)
    plt.setp(ax.get_yticklabels(), rotation=0, ha="center", fontsize=9)
    
    # Estética e layout
    ax.set_title("Backtest Fórmula Mágica", fontsize=14, fontweight="bold")
    ax.set_ylabel("Retorno Acumulado (Multiplicador)", fontsize=9)

    ax.grid(axis="y", linestyle="--", alpha=0.6)
    ax.grid(axis="x", which='major', linestyle='', alpha=0)

    ax.legend()
    fig.tight_layout()
    plt.show()
```

```python
df_resultados = const_port_FM(df_ROIC, df_EY, 1)
df_resultados.tail()
```

```python
backtesting_FM(df_ROIC, df_EY, [1, 5, 20, 50, 90])
```
